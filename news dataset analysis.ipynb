```python
dataset = pd.get_dummies(df, columns=['article_id'])
```

```python
for i in df.columns:
    print(i)
```

```python
dataset.head()
```

```python
col=[]
for columns in dataset.columns:
    if('article_id_' in columns):
        col.append(columns.replace('article_id_',''))
    else:
        col.append(columns)
```

```python
col
```

```python
df_new = dataset
df_new.columns = col
```

```python
df_new.head()
```

```python
df_new.to_csv('news_data_article_one_hot.csv',index=None)
```
